````python
import numpy as np
from scipy.linalg import lu
A = np.array(object=
             [[2,2,2],
              [4,7,7],
              [6,18,22]],dtype=np.float64)

B = np.array(object=[12,24,12],dtype=np.float64)

P,L,U = lu(A)
print('L = \n',L,'\n')
print('U = \n',U,'\n')
````

````python
y = np.linalg.solve(L,B) # solve equation : Ly = B
y
````

````python
x = np.linalg.solve(U,y) # solve equation : Ux = y
x
````

### <font color = green|><b>Exercise 05:</b>
$
a.)\space Describe\space the\space steps\space for\space computing\space the\space PCA\space (Principal\space Component\space Analysis)
$

<div style="display: flex; justify-content: center;">

 

<font color = green|>

$
b.)\space Write\space a\space code\space in\space Python\space to\space perform\space PCA\space by\space using\space the\space steps\space in\space (a).
$

````python
import numpy as np

def perform_pca(data, n_components):
    
    # Step 1: Standardize the data
    standardized_data = (data - np.mean(data, axis=0)) / np.std(data, axis=0)
    
    # Step 2: Compute the covariance matrix
    covariance_matrix = np.cov(standardized_data.T)
    
    # Step 3: Eigendecomposition of the covariance matrix
    eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
    
    # Step 4: Sort eigenvalues and eigenvectors
    sorted_indices = np.argsort(eigenvalues)[::-1]
    sorted_eigenvalues = eigenvalues[sorted_indices]
    sorted_eigenvectors = eigenvectors[:, sorted_indices]
    
    # Step 5: Select the desired number of principal components
    selected_eigenvectors = sorted_eigenvectors[:, :n_components]
    
    # Step 6: Construct the projection matrix
    projection_matrix = selected_eigenvectors
    
    # Step 7: Transform the data
    transformed_data = np.dot(standardized_data, projection_matrix)
    
    return transformed_data

# Example usage
data = np.array([[1, 2, 3, 0], 
                 [4, 5, 6, -1], 
                 [7, 8, 9, -4]])  # Replace with your own data
n_components = 3  # Replace with the desired number of principal components

transformed_data = perform_pca(data, n_components)
print(transformed_data)
````
