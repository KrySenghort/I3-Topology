## <font color = green|><b><p style='text-align:center'>Gradient Descent Algorithm 

$ 
Step\space01:\space Declare\space two\space set\space of\space feature. 
$
$$
eg: x = np.array(object = [x_1,x_2,x_3,...,x_n],dtype=[np.float\space,\space np.int64])
$$
$$
eg: y = np.array(object = [y_1,y_2,y_3,...,y_n],dtype=[np.float\space,\space np.int64])
$$
$
Step\space 02:\space Compute\space mean\space value\space and\space standard\space deviation\space of\space x 
$

$
Step\space 03:\space Plot\space the\space graph\space between\space x\space and\space y\space by\space using\space scatter\space plot 
$

$
Step\space 04:\space Define\space hypothesis\space function.\space where\space
$
$$
f(x) = y = ax+b
$$
$$
x\space:\space variable\space feature,\space scalar\space or\space numpy\space array.
$$
$$
a,b:\space scalar.
$$
$
Step\space 05:\space Define\space cost\space function. where\space 
$
$$
j = \frac{(f(x)-y)^{2}}{2n} 
$$
$
Step\space 06:\space Find\space gradient\space of\space cost\space function 
$
$$
grad(j) = \frac{(f(x)-y)x}{n}
$$

$
Step\space 07:\space Training\space data 
$

$
Step\space 08:\space Converge\space value\space of\space a\space and\space b
$

$
Step\space 09:\space Plot\space cost\space function\space vs\space Iteration 
$

$
Step\space 10:\space Evaluation\space the\space model  
$

$
Step\space 11:\space Predict\space mass\space of\space students
$

```python
import numpy as np 
import matplotlib.pyplot as plt 
x = np.array(object=[152,157,160,163,150,147,165,168,178]) # height of student in cm
y = np.array(object=[54.48,55.84,57.20,58.57,53.12,52.21,59.93,61.29,69.92]) # mass of student in kg
x_bar = np.mean(a=x)
x_sd = np.std(a=x)
x = (x-x_bar)/x_sd # feature scalling

# plot data 
plt.scatter(x,y)
plt.title('Height Vs Mass of Students')
plt.xlabel('height of student in cm')
plt.ylabel('mass of student in kg')
plt.show()

# defint hypothesis function 
# x: variable feature, scalar or numpy array.
# a,b: scalar.
def hyp(x,a,b):
    y = a*x + b 
    return y 

# Define cost function 
def cost_func(x,a,b,y):
    # cost function : 1/2 mean square error
    # y : target feature, numpy array
    j = ((hyp(x,a,b)-y)**2).mean()*1/2
    return j 

# Find gradient of cost function 
def grad_a(x,a,b,y):
    grad = ((hyp(x,a,b)-y)*x).mean()
    return grad 

def grad_b(x,a,b,y):
    grad = (hyp(x,a,b)-y).mean()
    return grad 

# Training data 
a,b = 0,0 # initialize a and b 
alpha = 0.01  # learning rate
iter = 1000
j_list = []
for iter in range(iter):
    j = cost_func(x,a,b,y)
    j_list.append(j) # add j to j_list 
    a += -alpha * grad_a(x,a,b,y)
    b += -alpha * grad_b(x,a,b,y)
    print(f'iter {iter} : mse = {j:.4f}')

# Converge value of a and b
print(f'a={a:.4f} , b = {b:.4f}')

# Plot Cost vs Iteration 
plt.scatter(range(iter+1),j_list)
plt.show()

# Evaluation 
RSS = ((hyp(x,a,b)-y)**2).sum()
TSS = ((hyp(x,a,b)-y.mean())**2).sum()
R_2 = 1 - (RSS/TSS)
print(f'R squared value is {R_2:.4f}')

# Predict mass of students
x_new = np.array([170,180,200])
x_new = (x_new-x_bar)/x_sd
y_predict = hyp(x_new,a,b)
plt.plot(np.linspace(-2,4,100),hyp(np.linspace(-2,4,100),a,b),color='blue')
plt.show()
print(x_new)
print(y_predict)
```
