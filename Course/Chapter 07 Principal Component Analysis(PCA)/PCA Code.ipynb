## <b><font color = green|>I.) Lecturer Follow

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
def pca(x,k):
    # substract the mean of each variable 
    x = x - np.mean(x,axis=0)
    
    # calculate the covariance matrix
    cov_matrix = np.cov(x.T)
    
    # calculate eigenvalues and eigenvectors of the covariance matrix 
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
    
    # sort the eigenvector and eigenvalues in descending order 
    sorted_indices = np.argsort(eigenvalues)[::-1]
    sorted_eigenvalues = eigenvalues[sorted_indices]
    sorted_eigenvectors = eigenvectors[:,sorted_indices]
    
    # select the first k eigenvectors
    principal_component = sorted_eigenvectors[:,:k]
    
    # project the data onto the principal components
    projected_data = x.dot(principal_component)
    return projected_data
```

## <font color = green|><b>Random Seed

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Generate random data for demonstration
np.random.seed(42)
X = np.random.rand(100, 2)  # 100 samples, 2 features

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Plot the original data and the transformed data after PCA
plt.scatter(X[:, 0], X[:, 1], label='Original Data')
plt.scatter(X_pca[:, 0], X_pca[:, 1], label='Transformed Data')

# Plot the principal components (eigenvectors)
origin = pca.mean_
components = pca.components_
scale = 3
for i, (var_ratio, component) in enumerate(zip(explained_variance_ratio, components)):
    
    plt.arrow(origin[0], origin[1], scale * var_ratio * component[0], scale * var_ratio * component[1],head_width=0.1, head_length=0.1, fc='r', ec='r')
    
    plt.text(scale * var_ratio * component[0] + origin[0] + 0.1,scale * var_ratio * component[1] + origin[1] + 0.1, f'PC{i+1}', fontsize=8)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('PCA Example')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for demonstration
np.random.seed(42)
X = np.random.rand(100, 2)  # 100 samples, 2 features

# Subtract mean from data
mean = np.mean(X, axis=0)
X_centered = X - mean

# Compute covariance matrix
covariance_matrix = np.cov(X_centered.T)

# Perform eigendecomposition on covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Sort eigenvalues and eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues_sorted = eigenvalues[sorted_indices]
eigenvectors_sorted = eigenvectors[:, sorted_indices]

# Select the top k eigenvectors based on explained variance ratio
k = 1
explained_variance_ratio = eigenvalues_sorted / np.sum(eigenvalues_sorted)
top_k_eigenvectors = eigenvectors_sorted[:, :k]

# Transform the data into the reduced-dimensional space
X_pca = np.dot(X_centered, top_k_eigenvectors)

# Plot the original data and the transformed data after PCA
plt.scatter(X[:, 0], X[:, 1], label='Original Data')
plt.scatter(X_pca[:, 0], np.zeros_like(X_pca[:, 0]), label='Transformed Data')

# Plot the principal component (eigenvector)
origin = mean
component = top_k_eigenvectors[:, 0]
scale = 3 * np.sqrt(explained_variance_ratio[0])

plt.arrow(origin[0], origin[1], scale * component[0], scale * component[1],head_width=0.1, head_length=0.1, fc='r', ec='r')
plt.text(scale * component[0] + origin[0] + 0.1,scale * component[1] + origin[1] + 0.1, 'PC1', fontsize=8)

print(X[:20])
print('\nAfter pca operation\n')
print(X_pca[:20])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('PCA Example')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Generate synthetic data
np.random.seed(42)
mean = [0, 0]
cov = [[1, 0.5], [0.5, 1]]
X = np.random.multivariate_normal(mean, cov, 100)

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the original data and the transformed data after PCA
plt.scatter(X[:, 0], X[:, 1], label='Original Data')
plt.scatter(X_pca[:, 0], X_pca[:, 1], label='Transformed Data')

# Plot the principal components (eigenvectors)
origin = [0, 0]
components = pca.components_
explained_variance = pca.explained_variance_
for i, (component, explained_var) in enumerate(zip(components, explained_variance)):
    
    scale = 3 * np.sqrt(explained_var)
    
    plt.arrow(origin[0], origin[1], scale * component[0], scale * component[1],head_width=0.1, head_length=0.1, fc='r', ec='r')
    plt.text(scale * component[0] + origin[0] + 0.1,scale * component[1] + origin[1] + 0.1, f'PC{i+1}', fontsize=8)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('PCA Example')
plt.legend()
plt.grid(True)
plt.show()
```

## <font color = green|><b>Iris Data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

print(X[:20])
print('\n')
print(X_pca[:20])

# Plot the transformed data
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA on Iris Dataset')
plt.colorbar(label='Target')
plt.grid(True)
plt.show()
```

## <font color = green|><b>Boston Housing Price Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the Boston Housing dataset
boston = load_boston()
X = boston.data
y = boston.target

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

np.set_printoptions(precision=6)
print(X[:10])
print('\n')
print(X_pca[:10])

# Plot the transformed data
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA on Boston Housing Dataset')
plt.colorbar(label='Price')
plt.grid(True)
plt.show()
```
