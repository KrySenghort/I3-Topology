```python
import numpy as np 
import matplotlib.pyplot as plt
```

<b><font color=yellow>np.linspace()</font></b> is a numPy function that is used to create an evenly spaced array of numbers over a specified interval. It takes three arguments:<br>
- start: the start of the interval
- stop: the end of the interval (inclusive)
- num: the number of points to generate<br>
np.linspace(start,stop,num)

```python
t = np.linspace(0,40,1000) # start, stop, step

# distance trvelled by Pisey : d = 2.5t
d_p = 2.5*t

# distance travelled by Dara
d_r = 3*(t-5)

fig,ax = plt.subplots()
plt.title('A Bank Pisey Meet')
plt.xlabel('time(in minutes)')
plt.ylabel('distance(in km)')

ax.set_xlim([0,40])   # set the limit mesurement on x-axis and y-axis
ax.set_ylim([0,100])

ax.plot(t,d_p,c='green')   
ax.plot(t,d_r,c='brown')

plt.axvline(x=30,color='purple',linestyle = '--')    # show the line of perpendicular of intersection point between d_r and d_p
plt.axhline(y=75,color='purple',linestyle='--')
plt.show()
```

```python
# Solving linear system
x = np.linspace(-10,10,1000) # start,stop,step
x
```

```python
y2 = 1 + (5*x)/2 
print(y2)
```

```python
fig,ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')

ax.set_xlim([0,3])
ax.set_ylim([0,8])

ax.plot(x,y1,c='green')
ax.plot(x,y2,c='brown')

plt.axvline(x=2,color='purple',linestyle = '--')
plt.axhline(y=6,color='purple',linestyle = '--')
plt.show()
```
