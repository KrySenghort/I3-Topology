```python
import numpy as np
A = np.array([[1,3,5],[2,4,6],[0,2,4]])
print('det(A) = ',np.linalg.det(A))
```

```python
B = np.array([[2,0,1,2,0],[2,-1,0,1,1],[0,1,2,1,2],[-2,0,2,-1,2],[2,0,0,1,1]])
print(np.linalg.det(B))
```

```python
C = np.matrix([[-2,2],[2,1]])
Eigenvalue, Eigenvector = np.linalg.eig(C)
Eigenvalue
```
